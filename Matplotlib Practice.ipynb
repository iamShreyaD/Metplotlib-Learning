```python
#grid function connects grid lines making plots easier to read by adding reference lines

bx = [1, 2, 3, 4, 5] 
by = [5, 10, 15, 20, 25]
plt.plot(bx, by)
plt.grid(axis="y",
        linewidth=2,
        color="lightgray",
        linestyle="dotted")
```

```python
#bar chart compares categories of data by representing each category with a bar

categories = ["Grains", "Fruit", "Vegetables", "Protein", "Dairy", "Sweets"]
values = np.array([4, 3, 2, 5, 3, 1])

#for verticale bar chart
plt.bar(categories, values, color="skyblue")

#for horizontal bar chart
#plt.barh(categories, values, color="skyblue")
plt.title("Daily Consumption")
plt.xlabel("Food")
plt.ylabel("Quantity")
```

```python
#Pie chart is a circular chart divided into slices to show percentages of the total. Good for visualizing distribution among categories.

categories = ['Freshman', 'Sophomores', 'Juniors', 'Seniors']
values = np.array([300, 250, 275, 225])
colors = ["red", "yellow","blue", "green"]

plt.pie(values, labels=categories,
       autopct="%1.1f%%",
       colors=colors,
       explode=[0,0,0,0.1],
       shadow=True,
       startangle=180)

plt.title("University of Toledo")
```

```python
#scatter graphs show the relationship between two variables and helps to identify a correlation (+, -, None)
#Example: Study hours vs. Test scores
cx = np.array([0, 1, 1, 2 ,3 , 4 , 5, 6, 7, 7, 8])  # hours studies
cy = np.array([55, 60, 65, 62, 68, 70, 75, 78, 82, 85, 87])   #grades

dx = np.array([0, 1, 2, 3, 3, 6, 5, 8, 10, 9, 8])  # hours studies
dy = np.array([55, 65, 65, 76, 68, 88, 98, 78, 92, 85, 87])   #grades

plt.scatter(cx, cy, color = "red",
           alpha=0.5,
           s=30,
           label="Class A")

plt.scatter(dx, dy, color = "green",
           alpha=0.5,
           s=30,
           label="Class B")

plt.xlabel("Hours Studied")
plt.ylabel("Grade")
plt.title("Test Scores")
plt.legend()
```

```python
#histogram is a visual representation of the distribution of quantitative data. It groups values into bins (intervals) and counts how many fall in each range.

scores = np.random.normal(loc=80, scale=10, size=100)
scores = np.clip(scores, 0, 100)
plt.hist(scores, bins=8,
        color="lightgreen",
        edgecolor="black")
plt.title("Exam Scores")
plt.xlabel("Score")
plt.ylabel("Number of Students")
```

```python
#figure is the entire canvas
#ax is a single plot (subplot)

ex = np.array([1, 2, 3, 4, 5])
figure, axes = plt.subplots(2, 2)

axes[0,0].plot(ex, ex*2, color="red")
axes[0,0].set_title("x*2")

axes[0,1].plot(ex, ex**2, color="blue")
axes[0,1].set_title("x**2")

axes[1,0].plot(ex, ex**3, color="green")
axes[1,0].set_title("x**3")

axes[1,1].plot(ex, ex**4, color="purple")
axes[1,1].set_title("x**4")

plt.tight_layout()
```
